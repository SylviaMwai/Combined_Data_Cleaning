```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("combined.csv")
data.head(8)
```

```python
data.info()
data.describe()
```

```python
# Remove duplicates and make the change permanent
data.drop_duplicates(inplace=True)
data.head()
```

```python
#filling missing values using the mean
#selects only the numeric columns and calculates the mean of each of those columns, and ignores the non-numeric ones.
data.fillna(data.select_dtypes(include=np.number).mean(), inplace=True) 
data.head()
```

```python
#missing values for non-numerics
#The lambda function applies the mode function to each non-numeric column and returns the most frequent value.
data.fillna(data.select_dtypes(exclude=np.number).apply(lambda x: x.mode()[0]), inplace=True)
data.head()
```

```python
#Data inconsistency
#str.strip - Removes leading and trailing whitespace.
#str.lower() - Converts all text to lowercase
#.replace - handles the dictionary format correctly and replaces the values as specified.
data['gender'] = data['gender'].str.strip().str.lower().replace({
    'male': 'Male',
    'm': 'Male', 
    'female': 'Female', 
    'f': 'Female'})
data.head()
```

```python
#Outliers
# Remove outliers based on 'hindfoot_length'
Q1_hindfoot = data['hindfoot_length'].quantile(0.25)
Q3_hindfoot = data['hindfoot_length'].quantile(0.75)
IQR_hindfoot = Q3_hindfoot - Q1_hindfoot
lower_bound_hindfoot = Q1_hindfoot - 1.5 * IQR_hindfoot
upper_bound_hindfoot = Q3_hindfoot + 1.5 * IQR_hindfoot

data = data[(data['hindfoot_length'] >= lower_bound_hindfoot) & (data['hindfoot_length'] <= upper_bound_hindfoot)]

# Remove outliers based on 'weight'
Q1_weight = data['weight'].quantile(0.25)
Q3_weight = data['weight'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight
lower_bound_weight = Q1_weight - 1.5 * IQR_weight
upper_bound_weight = Q3_weight + 1.5 * IQR_weight

data = data[(data['weight'] >= lower_bound_weight) & (data['weight'] <= upper_bound_weight)]
data
```

```python
Q1_year = data['year'].quantile(0.25)
Q3_year = data['year'].quantile(0.75)
IQR_year = Q3_year - Q1_year
lower_bound_year = Q1_year - 1.5 * IQR_year
upper_bound_year = Q3_year + 1.5 * IQR_year

outliers_year = data[(data['year'] < lower_bound_year) | (data['year'] > upper_bound_year)]
print(outliers_year)
```

```python
#Unwanted Variables
#make numbers to be the set index
data.set_index("number", inplace=True)
data
```

```python
#Data Formats
data['month'] = data['month'].astype(int) #converts it to integer
data['day'] = data['day'].astype(int) #converts it to integer
data['hindfoot_length'] = data['hindfoot_length'].round(3) #rounds it off to 3dp
data['weight'] = data['weight'].round(3) #rounds it off to 3dp
data.head()
```

```python
#exporting data
data.to_csv('combined_cleaned.csv', index=False)
```
